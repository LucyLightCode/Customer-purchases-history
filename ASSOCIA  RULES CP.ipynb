```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
# Toy dataset
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
               ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
               ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
               ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
               ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
te= TransactionEncoder()
te_ary= te.fit(toy_dataset).transform(toy_dataset) #To fit and encode the dataset
encoded_dataset= pd.DataFrame(te_ary, columns=te.columns_) #To return a dataframe of the encoded dataset
encoded_dataset
```

```python
support = apriori(encoded_dataset)
support #Put in the dataset in apriori
#It returns the support of each of the items or itemset. The items names are the indicies
```

```python
#to show the item names and the support
support_new = apriori(encoded_dataset,use_colnames=True)
support_new
```

```python
#for confidence and lift
from mlxtend.frequent_patterns import association_rules
association_rules(support_new, metric="confidence") # confidence - fit it in the dataset that that the support and itemset
```

```python
#for lift
association_rules(support_new,metric="lift")
```

Interpretation of result

There is a high support for hats, pants, scarfs and skirts of all the items in the store.

The confidence for Pants and Skirts is high

The lift for pants and skirt is 1.25 which means the two itmes go together. There is a high chance of a customer buy both items at the same time.

```python
Market = pd.read_csv('/content/drive/MyDrive/Market_Basket_Optimisation.csv', header = None)
Market.head()
```

```python
#To convert the dataframe to a nested list

Market_list = Market.values.tolist()
Market_list
```

```python
#To remove the Nan in the data
Market_list2 = [[x for x in y if pd.notna(x)] for y in Market_list]
Market_list2
```

```python
#Encode the new data and convert it to a dataframe
tr_en = TransactionEncoder()
Market_encoded =tr_en.fit(Market_list2).transform(Market_list2)
Market_encoded = pd.DataFrame(Market_encoded, columns=tr_en.columns_)
Market_encoded.head()
```

```python
#To get the support fit the encoded data set into apriori
support_1 = apriori(Market_encoded, min_support= 0.01)
support_1.head()
```

```python
Support_2 = apriori(Market_encoded, min_support=0.01, use_colnames=True)
Support_2.head()
```

```python
#for confidence
association_rules(Support_2, min_threshold= 0.01, metric="confidence")
```

```python
#for lift
association_rules(Support_2,min_threshold= 1, metric="lift")
```

Interpretation of Result

There is a high level of lift among most of the items. The supermarket should stock up on the items.
